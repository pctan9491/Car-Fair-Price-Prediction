````python
# create new feature (distance) and this is the calculation
import math
def haversine_distance(dep_long, dep_lat, arr_long, arr_lat):
    # Convert latitude and longitude from degrees to radians
    dep_long, dep_lat, arr_long, arr_lat = map(math.radians, [dep_long, dep_lat, arr_long, arr_lat])

    #Haversine Formula
    dlon = arr_long - dep_long
    dlat = arr_lat - dep_lat
    a=math.sin(dlat/2)**2+math.cos(dep_lat)*math.cos(arr_lat)*math.sin(dlon/2)**2
    c=2*math.asin(math.sqrt(a))
    radius=6371 #(Radius of the earth)
    return c*radius
````

````python
train_set["distance"]=train_set.apply(lambda row: haversine_distance(row['departure_long'], row['departure_lat'], row['arrival_long'], row['arrival_lat']), axis=1)
train_set["distance"]
````
